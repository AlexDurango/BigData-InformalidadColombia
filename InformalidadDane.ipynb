```python
# Hacer que las ciudades sean columnas y las fechas los indices
dictionaryNewDataframe = {}
for row in main_df.index.values.tolist():
    key = row
    key_column = []
    for column in main_df.columns.values.tolist():
        key_column.append(main_df[column][row])
    
    dictionaryNewDataframe[key] = key_column

new_df = pd.DataFrame(dictionaryNewDataframe)              
new_df.index = fechas_estudiadas
new_df.head()
```

```python
# Hacer el datafrane por el promedio de cada año

dictByYear = {}

for column in new_df.columns.values.tolist():
    lastYear = 2007
    sumatoria = 0
    elementosS = 0
    toAdd = []
    
    for idx in new_df.index.values.tolist():
        if int(idx.split()[0]) == lastYear:
            sumatoria += new_df[column][idx]
            elementosS += 1

        else:
            if elementosS != 0:
                meanOfYear = sumatoria / elementosS
                toAdd.append(meanOfYear)
            lastYear += 1
            sumatoria = 0
            elementosS = 0
    
    dictByYear[column] = toAdd

meanByYear_df = pd.DataFrame(dictByYear)
meanByYear_df.index = ['2007','2008','2009','2010','2011','2012','2013','2014',
                   '2015','2016','2017','2018','2019','2020']
meanByYear_df = pd.concat([meanByYear_df, pd.DataFrame(dict(new_df.loc['2021 Ene - Mar']), index=['2021'])], axis=0)
meanByYear_df
```

```python
# Como ha evolucionado la informalidad en el pais

x = [int(i) for i in meanByYear_df.index.values.tolist()]
y = [] # Array de informalidad por año

# Promedio de informalidad por año
for i in meanByYear_df.index.values.tolist():
    sumaValores = 0
    for j in meanByYear_df.columns.values.tolist():
        sumaValores += meanByYear_df[j][i]
        
    y.append(sumaValores/len(meanByYear_df.columns.values.tolist()))

plt.scatter(x,y,c=[[0,0,1]])
plt.xlabel('Años')
plt.ylabel('Proporción de informalidad')
plt.title("Evolución de la informalidad en Colombia")
plt.show()
```
